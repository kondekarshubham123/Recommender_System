```python
# Importing Libraries
import pandas as pd
import numpy as np
import pickle
from sklearn.metrics.pairwise import cosine_similarity

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading file
df = pd.read_csv('CCMLEmployeeData.csv')
```

```python
# Visualising file
sns.countplot(df.Domain)
plt.xticks(rotation=90)
```

```python
sns.countplot(df.Event1)
plt.xticks(rotation=90)
```

```python
sns.countplot(df.Event2)
plt.xticks(rotation=90)
```

```python
df.describe()
```

```python
df['Domain'] = df['Domain'].str.lower()
df['Event1'] = df['Event1'].str.lower()
df['Event2'] = df['Event2'].str.lower()
print(df.head())
```

```python
events = np.unique(df['Event1'].values)
dic = {}
for index,i in enumerate(events):
	dic[i] = index
```

```python
df['Event1'] = df['Event1'].map(dic)
df['Event2'] = df['Event2'].map(dic)
```

```python
pickle.dump(dic,open('Model/Events.pickle','wb+'))
```

```python
domains = np.unique(df['Domain'].values)
```

```python
domain = {}
for index,i in enumerate(domains):
	domain[i] = index
```

```python
df['Domain'] = df['Domain'].map(domain)
```

```python
pickle.dump(domain,open('Model/Domain.pickle','wb+'))
print(df['Domain'].head())
df.to_csv('Model/preprocessing.csv')
```
